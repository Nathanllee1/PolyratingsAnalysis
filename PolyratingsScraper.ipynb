```python
from bs4 import BeautifulSoup
import requests
```

```python
allProfessors = requests.get('http://polyratings.com/list.php').text
```

```python
page = BeautifulSoup(allProfessors)
```

```python
profObjects = page.find_all(class_='no-link-highlight text-muted filterable')
```

```python
linkList = []

for links in profObjects:
    linkList.append(links['href'])
```

```python
print(linkList)
```

```python
import pymongo
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient['mydatabase']
mycol = mydb["data"]
```

```python
dataObj = {teacher, description, course, year, grade, choice, year, comment}
```

```python
dataList = []
for links in linkList:
    try:
        professorReq = requests.get(links).text
    except:
        pass
    professorPage = BeautifulSoup(professorReq)
    idNum = links.split("=")[1]
    print(idNum)
    teacher = professorPage.find(class_='text-primary').strong.get_text()
    
    groups = professorPage.find_all('section', class_='group')
    for classSection in groups:
        course = classSection.find('h2').text
        commentObjects = classSection.find_all(class_="col-xs-12 item-padding")
        
        for comments in commentObjects:
            infoColumn = comments.find(class_='col-xs-3 col-sm-2 eval-info').text.split('\n')
            gradeLevel = infoColumn[1]
            grade = infoColumn[2]
            choice = infoColumn[3]
            date = infoColumn[4]
            
            comment = comments.find(class_="col-xs-9 col-sm-10 eval-comment").text
            
            dataList.append({'idNum':idNum, 'teacher':teacher, 'course':course, 'gradeLevel':gradeLevel, 'grade':grade, 'choice':choice, 'date':date, 'comment':comment})
            
            
    print(teacher)
```

```python
import pandas
```

```python
df = pandas.DataFrame(dataList)
```
